```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
import pydot

SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)
```

```python
from tensorflow.keras.preprocessing.image import load_img
```

```python
load_img('data/train/curly/00cbad1ffe22d900018e5a2e7376daed4.jpg')
```

```python
def make_model_act(input_shape):
    inputs = keras.Input(shape=input_shape)
    layer_1 = keras.layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu')(inputs)
    layer_2 = keras.layers.MaxPooling2D(pool_size=(2, 2))(layer_1)
    vectors = keras.layers.Flatten()(layer_2)
    layer_3 = keras.layers.Dense(units=64, activation='relu')(vectors)
    outputs = keras.layers.Dense(units=1, activation='sigmoid')(layer_3)

    model = keras.Model(inputs, outputs)

    #############################

    optimizer = keras.optimizers.SGD(learning_rate=0.002, momentum=0.8)

    loss = keras.losses.BinaryCrossentropy() 

    model.compile(
        optimizer=optimizer, 
        loss=loss, 
        metrics=['accuracy']
        )

    return model
```

```python
input_shape=(200, 200, 3)
model = make_model_act(input_shape=input_shape)
```

```python
model.summary()
```

```python
keras.utils.plot_model(
    model,
    show_shapes=True,
    show_layer_names=True,
    rankdir="TB",
    dpi=100,
    show_layer_activations=True,
    show_trainable=True,
)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
# from tensorflow.keras.applications.xception import Xception
# from tensorflow.keras.preprocessing.image import preprocess_input
# from tensorflow.keras.preprocessing.image import decode_predictions
```

```python
image_gen = ImageDataGenerator(rescale=1./255)
image_size = (200, 200)
```

```python
train_gen = image_gen.flow_from_directory(
    './data/train', 
    target_size = image_size, 
    class_mode='binary',
    batch_size=20,
    shuffle=True
    
) 

test_gen = image_gen.flow_from_directory(
    './data/test', 
    target_size = image_size, 
    class_mode='binary',
    batch_size=20,
    shuffle=True
) 
```

```python
history = model.fit(
    train_gen,
    epochs=10,
    validation_data=test_gen
)
```

```python
np.median(history.history['accuracy'])
```

```python
np.std(history.history['loss'])
```

```python
new_image_gen = ImageDataGenerator(
    rotation_range=50,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    fill_mode='nearest'
)

new_train_gen = new_image_gen.flow_from_directory(
    './data/train', 
    target_size = image_size, 
    class_mode='binary',
    batch_size=20,
    shuffle=True
    
) 
```

```python
history2 = model.fit(
    new_train_gen,
    epochs=10,
    validation_data=test_gen
)
```

```python
np.mean(history2.history['val_loss'])
```
